```python
import pandas as pd
import numpy as np
import folium
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/content/drive/MyDrive/Motor_Vehicle_Collisions_-_Crashes.csv')
df.head()  # View the first few rows
```

```python
df.isnull().sum()
```

```python
# Drop rows where LATITUDE or LONGITUDE are missing
df.dropna(subset=['LATITUDE', 'LONGITUDE'], inplace=True)

# Check if there are still missing values
df.isnull().sum()
```

```python
# Fill missing values in 'BOROUGH' with the most frequent value (mode)
df['BOROUGH'].fillna(df['BOROUGH'].mode()[0], inplace=True)

# Fill missing values in 'ZIP CODE' with the most frequent value (mode)
df['ZIP CODE'].fillna(df['ZIP CODE'].mode()[0], inplace=True)

# Check if there are still missing values
df.isnull().sum()
```

```python
df[['ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME']].isnull().sum()
```

```python
# Fill missing values in street name columns with 'Unknown'
df['ON STREET NAME'].fillna('Unknown', inplace=True)
df['CROSS STREET NAME'].fillna('Unknown', inplace=True)
df['OFF STREET NAME'].fillna('Unknown', inplace=True)

# Check if there are still missing values
df.isnull().sum()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Selecting the LATITUDE and LONGITUDE columns for clustering
coords = df[['LATITUDE', 'LONGITUDE']]
```

```python
# Standardizing the data (normalization)
scaler = StandardScaler()
coords_scaled = scaler.fit_transform(coords)
```

```python
# Applying K-Means clustering (for example, 5 clusters)
kmeans = KMeans(n_clusters=5, random_state=0)
df['Cluster'] = kmeans.fit_predict(coords_scaled)
```

```python
# Check the first few rows to see the cluster assignment
df[['LATITUDE', 'LONGITUDE', 'Cluster']].head()
```

```python
import folium
```

```python
# Initialize a base map (centered around NYC coordinates)
map_center = [40.7128, -74.0060]  # Approximate center of NYC
mymap = folium.Map(location=map_center, zoom_start=12)
```

```python
# Plot each point with a different color for each cluster
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        radius=5,
        color=f"#{hex(255 - row['Cluster'] * 50)[2:]}0000",  # Assign color based on cluster
        fill=True,
        fill_color=f"#{hex(255 - row['Cluster'] * 50)[2:]}0000",
        fill_opacity=0.6
    ).add_to(mymap)
```
